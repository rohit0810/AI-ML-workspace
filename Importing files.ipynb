## Web Scraping

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())  
```

```python
soup.find_all('h1')[0].text
```

```python
company=soup.find_all('div',class_='companyCardWrapper__primaryInformation')
```

```python
len(company)
```

```python
company
```

```python
name=[]
rating=[]
reviews=[]

for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('span',class_='companyCardWrapper__companyRatingValue').text.strip())
    
    

  
```

```python
name
```

```python
campany2=soup.find_all('div',class_='companyCardWrapper__tertiaryInformation')
```

```python
for i in campany2:
    #name.append(i.find('h2').text.strip())
    #rating.append(i.find('span',class_='companyCardWrapper__companyRatingValue').text.strip())
    reviews.append(i.find('span',class_='companyCardWrapper__ActionCount').text.strip())
```

```python
reviews
```

```python
list_of_tuples = list(zip(name, rating, reviews))
```

```python
df7 = pd.DataFrame(list_of_tuples,
                  columns=['Name', 'rating', 'reviews'])
df7
```
