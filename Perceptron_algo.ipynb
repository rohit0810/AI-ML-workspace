```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_nums_1 = np.random.randint(10, 51, size=100)
random_nums_2 = np.random.randint(10, 51, size=100)
third_column = np.random.randint(0, 2, size=100)
df = pd.DataFrame({
    'Col1': random_nums_1,
    'Col2': random_nums_2,
    'target': third_column
})

print(df)
```

```python
# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(df['Col1'], df['Col2'], c=df['target'], cmap='coolwarm')
plt.xlabel('Column_1')
plt.ylabel('Column_2')
plt.title('Scatter Plot of Column_1 vs Column_2 with Column_3 as Labels')
plt.colorbar(label='Column_3')
plt.grid(True)
plt.show()
```

```python
def perceptron(df):
    weight = np.array([0, 1, 1])
    for i in range(200):
        stud = np.random.randint(0, len(df))  # Random index within the DataFrame range
        row_values = df.iloc[stud, :2].values
        x = np.insert(row_values, 0, 1)
        y_pred = np.dot(weight, x)
        if y_pred >= 0:
            y_pred = 1
        else:
            y_pred = 0
        weight = weight + 0.01 * (df['target'].iloc[stud] - y_pred) * x
    return weight[0], weight[1:]

p, q = perceptron(df)
print("Bias:", p)
print("Weights:", q)
```
