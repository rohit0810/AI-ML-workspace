```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
import pandas as pd
```

```python
df=pd.read_csv("E:/excel/sal/Salaries.csv")
```

```python
#We are interested only in Total Pay column
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], kde=True, stat="density", linewidth=0)
plt.title('Probability Density Function (PDF) of TotalPay')
plt.xlabel('TotalPay')
plt.ylabel('Density')
```

```python
sample_dist=[]
for i in range(3000):
    sample_dist.append(np.mean(df['TotalPay'].sample(n=30)))
```

```python
sample_dist
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(sample_dist, kde=True, stat="density", linewidth=0)
plt.title('Probability Density Function (PDF) of Sample Distribution TotalPay')
plt.xlabel('TotalPay')
plt.ylabel('Density')
```

```python
#comparing mean and variance
print("Mean of TotalPay Column of Original Data is :",np.mean(df['TotalPay']))
print("Mean of TotalPay Column of Sample Distribution is :",np.mean(sample_dist))
print("Variance of TotalPay Column of Original Data is :",df['TotalPay'].var())
print("Variance of TotalPay Column of Sample Distribution is :",statistics.variance(sample_dist))
```

```python
#We can verify that variance of the sample distribution is variance of TotalPay Column of original Data divided by sample size
df['TotalPay'].var()/30
```
